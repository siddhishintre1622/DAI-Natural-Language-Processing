```python
from nltk.tokenize import  word_tokenize
import pandas as pd
import nltk
nltk.download('punkt_tab')
```

```python
#Creating Bag of Words manually

sent1 = "It is a good practice for us."
sent2 = "It was also good to know about it."

sent = sent1 + " " + sent2
words = list(set(word_tokenize(sent.lower())))

s1 = []
s2 = []

for word in words:
    s1.append(word_tokenize(sent1.lower()).count(word))

for word in words:
    s2.append(word_tokenize(sent2.lower()).count(word))

print(words)
print(s1)
print(s2)

bog = pd.DataFrame({'words' : words})
bog['sent1'] = s1
bog['sent2'] = s2

df =  pd.DataFrame({},index=['sent1','sent2'], columns=bog['words'])
df.iloc[0,:] = s1
df.iloc[1,:] = s2

df
```

**Using in-built library**

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cvt = CountVectorizer()
x_new = cvt.fit_transform([sent1,sent2])

x_new.toarray()
```

```python
cvt.get_feature_names_out()
```

```python
df = pd.DataFrame(data=x_new.toarray(),columns=cvt.get_feature_names_out())
```

```python
df
```

```python
new = 'It was good for us'
new_features = cvt.transform([new])
new_features.toarray()
```
