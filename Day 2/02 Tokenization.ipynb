```python
sents = '''Hello friends!
How are you? Welcome to the world of Python Programming.'''
```

```python
#count percentage of punctuations in it
tokens = word_tokenize(sents)
```

1st Method

```python
cnt=0
for i in tokens:
  if i.isalnum():
    pass
  else:
    cnt+=1

print((cnt/len(tokens)*100))
```

2nd Method

```python
import string
string.punctuation
```

```python
count = [word for word in tokens if word in string.punctuation]
```

```python
len(count) / len(tokens) * 100
```

3rd Method

```python
count = [word for word in tokens if not word.isalpha() and not word.isdigit()]
```

```python
len(count) / len(tokens) * 100
```

MWE Tokenizer

```python
from nltk.tokenize import MWETokenizer
data=  'Van Rossom is in Pune today. We welocomed Van Rossom here.'
tk = MWETokenizer()
tk.add_mwe(('Van', 'Rossom'))
tk.tokenize(word_tokenize(data))
```

```python
tk= MWETokenizer(separator=' ')
tk.add_mwe(('Van', 'Rossom'))
tk.tokenize(word_tokenize(data))
```

Custom Tokenizer

```python
import re

def custom_tokenizer(text):
  return re.split(r"[.,;?!\s]+", text)

text = "This is some text with puctuation > Let's tokenize it. Is it ok?"

tokens = custom_tokenizer(text)

print("Tokens:")
for token in tokens:
  print(token)
```
