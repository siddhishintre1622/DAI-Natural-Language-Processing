```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('indian')
```

**Punctuation Removal**

```python
text = 'Hello friends! How are you? Welcome to Pune.'
for i in word_tokenize(text):
    if i.isalpha() or i.isdigit():
        print(i, end=' ')
```

```python
text = 'Hello friends! How are you? Welcome to Pune.'
text1=""
for i in word_tokenize(text):
    if i.isalpha() or i.isdigit():
        text1= text1 + i + ' '
print(text1)
```

```python
# List comprehension and string.punctuation

import string
tokens = [word for word in word_tokenize(text) if word not in string.punctuation]
text = ' '.join(tokens)
text
```

**Stopwords Removal**

```python
swords = stopwords.words('english')
swords
```

```python

text = ' '.join(tokens)
words = [word for word in word_tokenize(text) if word not in swords]
print(words)
#to remove 'how' stopword
text = ' '.join(tokens)
words = [word for word in word_tokenize(text) if word.lower() not in swords]
words
```
