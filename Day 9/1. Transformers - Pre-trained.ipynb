```python
from transformers import pipeline  #Used for creating an NLP pipeline for tasks like sentiment analysis, text generation, etc.
import pandas as pd
```

```python
classifier = pipeline('text-classification', model='t5-base')
```

```python
text = 'It was possible for him easily. So he did it.'
result = classifier(text)   #object
```

```python
df = pd.DataFrame(result)
df
```

```python
text = 'Nothing is possible without these efforts.'
result = classifier(text)
df = pd.DataFrame(result)
df
```

```python
text = 'I think it is impossible for him. So just avoid.'
result = classifier(text)
df = pd.DataFrame(result)
df
```

```python
text = 'It was so impressive but not upto the mark.'
result = classifier(text)
df = pd.DataFrame(result)
df
```

```python
text = 'You’re great at optimizing processes, devising new programs to improve the customer experience, and you love collaborating and developing your team.'
result = classifier(text)
df = pd.DataFrame(result)
df
```

**Question - Answering**

```python
qa = pipeline('question-answering')
```

```python
text = 'India, officially the Republic of India,[j][20] is a country in South Asia. It is the seventh-largest country in the world by area and the most populous country. Bounded by the Indian Ocean on the south, the Arabian Sea on the southwest, and the Bay of Bengal on the southeast, it shares land borders with Pakistan to the west;[k] China, Nepal, and Bhutan to the north; and Bangladesh and Myanmar to the east. In the Indian Ocean, India is in the vicinity of Sri Lanka and the Maldives; its Andaman and Nicobar Islands share a maritime border with Thailand, Myanmar, and Indonesia.'
```

```python
question = 'What is in the south of India?'
result = qa(question = question, context = text)
df = pd.DataFrame([result])
df
```

```python
question = 'India?'
result = qa(question = question, context = text)
df = pd.DataFrame([result])
df
```

```python
question = 'What is in the north of India?'
result = qa(question = question, context = text)
df = pd.DataFrame([result])
df
```

```python
question = 'Which is most populous country?'
result = qa(question = question, context = text)
df = pd.DataFrame([result])
df
```

***Visual Question Answering***

```python
from PIL import Image
```

```python
vqa = pipeline('visual-question-answering')
```

```python
image = Image.open('dog.jpg')
```

```python
question = 'What is color of dog?'
vqa(image, question, top_k=3) [0] ['answer']
```

```python
question = 'What dog is doing?'
vqa(image, question, top_k=3) [0] ['answer']
```

```python
question = 'What is the breed of dog?'
vqa(image, question, top_k=3) [0] ['answer']
```

```python
question = 'What is the name of dog?'
vqa(image, question, top_k=3) [0] ['answer']
```

```python
question = 'Where is the dog sitting?'
vqa(image, question, top_k=3) [0] ['answer']
```

```python
question= 'is the dog cute?'
vqa(image, question, top_k=3) [0] ['answer']
```
