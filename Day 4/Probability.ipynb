Theoritical Probability

```python
def coin_flip_probability():
    heads_probability = 1/2
    return heads_probability

def rolling_dice_probability(number):
    if number not in range(1,7):
        raise ValueError("Number must be between 1 and 6")
    probability = 1/6
    return probability
```

```python
#Example Usage
coin_flip_result  = coin_flip_probability()
dice_roll_result = rolling_dice_probability(4)

print("Probability of heads in a coin flip:", coin_flip_result)
print("Probability of rolling a 4 on a dice:", dice_roll_result)
```

```python
#Example Usage
dice_roll_result = rolling_dice_probability(4)
print("Probability of rolling dice:", dice_roll_result)
```

Experimental Probability

```python
import random
import numpy as np
```

```python
def simulate_coin_flips(num_flips):
    results = []
    for _ in range(num_flips):
        flip = random.choice(["heads", "tails"])
        results.append(flip)
    return results
```

```python
simulate_coin_flips(10)
```

To count

```python
import pandas as pd
pd.Series(simulate_coin_flips(1000)).value_counts()    #1st method
```

```python
from collections import Counter  #2nd method
Counter(simulate_coin_flips(1000))
```

```python
def calculate_experimental_probability(results, event):
    event_count = results.count(event)
    total_flips = len(results)
    probability = event_count / total_flips
    return probability
```

```python
#Example usage
num_flips = 1000
simulation_results = simulate_coin_flips(num_flips)

heads_probability = calculate_experimental_probability(simulation_results, "heads")
tails_probability = calculate_experimental_probability(simulation_results, "tails")

print(f"Number of flips: {num_flips}")
print(f"Heads Probability: {heads_probability:.4f}")
print(f"Tails Probability: {tails_probability:.4f}")
```

```python
heads_probability + tails_probability
```

Linear Growth

```python
x = np.arange(1,10,0.25)
```

```python
x
```

```python
y = 5.67 * x + 0.356  #mx+c
y
```

```python
plt.plot(x,y)
plt.title("Linear Growth")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
y0 = -5.67 * x + 0.356      #decreasing
plt.plot(x,y0)
plt.title("Linear Growth")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

Polynomial/Exponential Growth

```python
y2 = x** 2
plt.plot(x,y2)
plt.title("Polynomial Growth")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

Logarithmic Growth

```python
plt.plot(x, np.log(x))
plt.title("Logarithmic Growth")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

Logistic Growth

```python
x = np.arange(-5,5,0.25)
```

```python
x
```

```python
def sigmoid(z):
    val = 1/(1+np.exp(-z))
    return val
```

```python
sigmoid(x)
```

```python
plt.plot(x,sigmoid(x))
plt.title("Logistic Growth")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
