```python
#Reading data from pdf

from pypdf import PdfReader
reader = PdfReader('course.pdf')    #Creating object and passing pdf

```

```python
print(reader.pages[0])         # reading page 0
print(len(reader.pages))       # Finding length
```

```python
#Reading data from page
page1 = reader.pages[0]
page1 = page1.extract_text()
print(page1)
```

```python
#Reading complete document
for page in reader.pages:
    print(page.extract_text())
```

```python
# Saving the data in course.txt using file handling

with open('course.txt', 'w') as f:
    for page in reader.pages:
        f.write(page.extract_text())
```

```python
# saving data in 'text' and then writing in file
text = ''
for page in reader.pages:
    text+=page.extract_text()

file = open('course.txt', 'w')
file.write(text)
file.close()
```

**Working with DOCX**

```python
from spire.doc import Document
```

```python
document = Document()                     # Creating object
document.LoadFromFile('SPP.docx')         # Loading data
text = document.GetText()                 # Reading text
print(text)
```

```python
print(document.GetPageCount())
```

```python
section  = document.Sections[0]
print(section)
```

```python
paragraph = section.Paragraphs[1]
print(paragraph.Text)
```

```python
# Copying document page by page
from spire.doc import FixedLayoutDocument
```

```python
layoutDoc = FixedLayoutDocument(document)  # Creating object and passing document
page0 = layoutDoc.Pages[0]                 # Extracting page
```

```python
print(len(layoutDoc.Pages))  #Toal no of pages
```

```python
print(page0.Text)   #printing text of page
```
