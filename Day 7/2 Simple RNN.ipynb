This is multicalss classification and classes are 25 A to Z.
op = softmax
rnn = tanh

```python
alphabets= 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
for index,element in enumerate(alphabets):
    print(index , " : " , element)
```

```python
int_to_char = dict(enumerate(alphabets))
int_to_char[5]
```

```python
char_to_int = dict((v,i) for i,v in enumerate(alphabets))
char_to_int['A']
```

```python
seq_len= 1
x=[]
y=[]

for i in range(len(alphabets) - seq_len):
    seq_in =  [alphabets[i]]
    seq_out = alphabets[i+1]
    print(seq_in , ":" , seq_out)
    x += [[(char_to_int[seq_in[0]])]]
    y += [(char_to_int[seq_out])]

```

```python
x
```

```python
y
```

**Prpare Input Data**

```python
import numpy as np
```

```python
x = np.reshape(x , (len(x) , 1 , 1))
```

```python
x
```

```python
x.shape
```

```python
x = x/25  # Normalizing by dividing max no 25,25 is not present in train but present in test.  same as MinMaxScaler()
```

```python
x
```

**Prepare Output Data**

```python
from keras.utils import to_categorical
```

```python
y_new = to_categorical(y)
y_new.shape
```

```python
y_new
```

**Build the model**

```python
from keras.models import Sequential
from keras.layers import Dense , SimpleRNN,Input
```

```python
model = Sequential()
model.add(Input((1,1)))
model.add(Dense(25,activation='relu'))
model.add(SimpleRNN(32))
model.add(Dense(26,activation='softmax'))
```

```python
model.summary()
```

```python
from keras.utils import plot_model
```

```python
plot_model(model, show_layer_activations=True,
           show_shapes=True, show_layer_names=True)
```

```python
# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#Training the model
model.fit(x,y_new,epochs=1000,batch_size=1)
```

```python
model.evaluate(x,y_new,batch_size=1)
```

**Increasing Sequence length**

```python
seq_len= 3
x=[]
y=[]

for i in range(len(alphabets) - seq_len):
    seq_in =  alphabets[i:i+seq_len]
    seq_out = alphabets[i+seq_len]
    x.append([char_to_int[char] for char in seq_in])
    y.append(char_to_int[seq_out])
    print(seq_in , ":" , seq_out)
```

```python
x
```

```python
y
```

```python
x_new = np.reshape(x,(23,seq_len,1))
x_new = x_new/25
x_new
```

```python
y_new = to_categorical(y)
y_new.shape
```

```python
model = Sequential()
model.add(Input((3,1)))
model.add(Dense(25,activation='relu'))
model.add(SimpleRNN(32))
model.add(Dense(26,activation='softmax'))
```

```python
model.summary()
```

```python
plot_model(model, show_layer_activations=True,
           show_shapes=True, show_layer_names=True)
```

```python
# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#Training the model
model.fit(x_new,y_new,epochs=1000,batch_size=1)
```

```python
model.evaluate(x_new,y_new,batch_size=1)
```
